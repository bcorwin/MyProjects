```python
#Generate squares
sqs = [i**2 for i in range(100000)]
dsqs = [2*i for i in sqs]
```

```python
n = 100000
ans = {(a,b,c):False for c in range(2, n+1, 2) for b in range(int(c/2)+1, c+1) for a in range(c - b + 1 + (c+1)%2 ,b+1, 2)}
```

```python
# Rules:
## a == b (mod 2)
## c == 0 (mod 2)
## c/2 < b <= c
## c-b < a <= b
## a != c <- Not yet implemented
```

```python
def intergal_median(n):
    cnt = 0
    max_val = dsqs[-1]
    for tri,val in sorted(ans.items(), key = lambda ((a,b,c), value): (c,b,a)):
        if ans[tri] == True: continue
        
        chk = sqs[tri[0]] + sqs[tri[1]] - sqs[tri[2]]/2
        
        if chk in dsqs:
            max_k = int(n/tri[2])
            cnt += max_k
            for k in range(1,max_k+1):
                new = tuple([k*i for i in tri])
                #print new
                ans[new] = True
        elif chk > max_val: return("Error: need more squares")
    return(cnt)
```

```python
print(intergal_median(n))
```
